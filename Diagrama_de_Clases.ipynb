````python
class Cartelera:
    def __init__(self, Programacion, Preventa):
        self.Programacion = Programacion
        self.Preventa = Preventa

class Pelicula:
    def __init__(self, Nombre, Duracion, Clasificacion, Genero):
        self.Nombre = Nombre
        self.Duracion = Duracion
        self.Clasificacion = Clasificacion
        self.Genero = Genero

class Sala(Pelicula):
    def __init__(self, Numero_de_Sala, Tipo_de_Sala, Formato):
        self.Numero_de_Sala = Numero_de_Sala
        self.Tipo_de_Sala = Tipo_de_Sala
        self.Formato = Formato

class Cine(Cartelera, Sala):
    def __init__(self, Empresa, Direccion, Ciudad, n_de_asientos, n_Salas):
        self.Empresa = Empresa
        self.Ciudad = Ciudad
        self.Direccion = Direccion
        self.n_de_asientos = n_de_asientos
        self.n_Salas = n_Salas

    def __str__(self):
        return f" Empresa {self.Empresa}, Dirección {self.Direccion}, número de asientos {self.n_de_asientos}, numero de salas {self.n_Salas}"

class Promociones:
    def __init__(self, Cumpleaños, Descuentos_tarjetas, Descuento_empresa):
        self.Cumpleaños = Cumpleaños
        self.Descuentos_tarjetas = Descuentos_tarjetas
        self.Descuento_empresa = Descuento_empresa


class Medio_de_pago(Promociones):
    def __init__(self, Valor, Datos_Tarjeta):
        self.Valor = Valor
        self.Datos_Tarjeta = Datos_Tarjeta

class Venta(Medio_de_pago):
    def __init__(self, Presencial, Web):
        self.Presencial = Presencial
        self.Web = Web

class Usuario(Cine, Venta):
    def __init__(self, Nombre, Edad, Fecha_nacimiento, correo, Socio):
        self.Nombre = Nombre
        self.Edad = Edad
        self.Fecha_nacimiento = Fecha_nacimiento
        self.correo = correo
        self.Socio = Socio
````
